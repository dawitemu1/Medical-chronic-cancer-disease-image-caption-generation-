```python
import os
import shutil

# Path to the directory containing images
image_dir = 'D:/PhD file/image caption/PhD datasets/cancer_image with caption'

# Path to the file containing caption entries
caption_file_path = 'D:/PhD file/image caption/PhD datasets/caption.txt'

# Path to the directory where you want to save matching images and captions
output_dir = 'D:/PhD file/image caption/PhD datasets/cancer_caption'
os.makedirs(output_dir, exist_ok=True)

# Extract all unique image names from the caption file
image_names_in_caption = set()

with open(caption_file_path, 'r') as caption_file:
    for line in caption_file:
        if line.strip():  # Avoid processing empty lines
            # Split the line to extract the image name before the "#"
            image_name_with_ext = line.split()[0].split('#')[0]
            image_names_in_caption.add(image_name_with_ext)

# Iterate over all files in the image directory
for filename in os.listdir(image_dir):
    # Get the image name without the extension
    image_name, image_ext = os.path.splitext(filename)
    
    # Check if this image name is present in the caption file
    if filename in image_names_in_caption:
        # Copy the image to the output directory
        image_path = os.path.join(image_dir, filename)
        shutil.copy(image_path, output_dir)
        print(f"Copied image: {image_path} to {output_dir}")

print("complete.")
```
